```python
import tensorflow as tf 
import numpy as np
```

```python
# x,y train data 셋팅
x_train=[1,2,3,4]
y_train=[0,-1,-2,-3]
```

```python
# 모델 준비
tf.model = tf.keras.Sequential()
```

```python
# 모델 keras layer 더하기
tf.model.add(tf.keras.layers.Dense(units=1, input_dim=1))
```

```python
# sigmoid 값 설정
sgd = tf.keras.optimizers.SGD(lr=0.1)
tf.model.compile(loss='mse', optimizer=sgd)

tf.model.summary()
```

```python
tf.model.fit(x_train, y_train, epochs=200)
```

```python
y_prediction = tf.model.predict(np.array([5,4]))
print(y_prediction)
```
