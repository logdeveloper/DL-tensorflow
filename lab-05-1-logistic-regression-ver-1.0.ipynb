```python
# import tensorflow as tf 
# tensorflow 2.3 version 말고 1.0 version 사용
import tensorflow.compat.v1 as tf 
tf.disable_v2_behavior()

import numpy as np
```

```python
tf.set_random_seed(777)

# x1, x2
x_data=[[1,2],
       [2,3],
       [3,1],
       [4,3],
       [5,3],
       [6,2]]

# binary 값, 0이나 1의 값을 가짐.
y_data=[[0],
       [0],
       [0],
       [1],
       [1],
       [1]]

X= tf.placeholder(tf.float32, shape=[None,2])
Y= tf.placeholder(tf.float32, shape=[None,1])

# [2.1] x : 2개,  y : 1개 
W = tf.Variable(tf.random_normal([2,1]), name='weight')
# [1] y의 나가는 값이 1개 이므로 
b = tf.Variable(tf.random_normal([1]), name='bias')

# exp수식을 직접 이용해도 되지만 sigmoid 활용하여 hypothesis 선언한다.
# exp 수식 이용 : tf.div(1., 1. + tf.exp(tf.matmul(X,W)))
hypothesis = tf.sigmoid(tf.matmul(X,W) + b) 

cost = -tf.reduce_mean(Y * tf.log(hypothesis) + (1-Y) * tf.log(1-hypothesis))

# 직접 미분할 필요없고, tensorflow를 통해 계산하면 된다.
train = tf.train.GradientDescentOptimizer(learning_rate=0.01).minimize(cost)

# 0.5를 기준으로 0.5보다 크면 1로 측정함.
predicted = tf.cast(hypothesis > 0.5, dtype=tf.float32)
accuracy = tf.reduce_mean(tf.cast(tf.equal(predicted, Y), dtype=tf.float32))

# 실행
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    
    for step in range(10001):
        cost_val, _ = sess.run([cost, train], feed_dict={X:x_data, Y : y_data})
        if step % 500 ==0:
            print(step, cost_val)
            
            
    h, c, a = sess.run([hypothesis, predicted, accuracy],
                      feed_dict={X:x_data, Y:y_data})
    
    print("\nHypothesis: ", h, "\nCorrect (Y): ", c, "\nAccuracy: ",a)
```

cost는 점점 줄어들고, Y예측값과 동일하게 예측한 것을 확인할 수 있다.

실제 사례 적용 : 당뇨병 예측하기 (classifying diabets)

```python
tf.set_random_seed(777)

# 실데이터 불러오기
xy = np.loadtxt('../../../DeepLearningZeroToAll/DeepLearningZeroToAll-master/data-03-diabetes.csv', delimiter=',', dtype=np.float32)
x_data = xy[:, 0:-1]
y_data = xy[:, [-1]]

print(x_data.shape, y_data.shape)
```

```python
X = tf.placeholder(tf.float32, shape=[None, 8])
Y = tf.placeholder(tf.float32, shape=[None, 1])

# x가 8개 y는 1개
W = tf.Variable(tf.random_normal([8,1]), name = 'weight')
b = tf.Variable(tf.random_normal([1]), name = 'bias')

hypothesis = tf.sigmoid(tf.matmul(X,W) + b)

cost = -tf.reduce_mean(Y * tf.log(hypothesis) + (1-Y)*tf.log(1-hypothesis))
train = tf.train.GradientDescentOptimizer(learning_rate=0.01).minimize(cost)

predicted = tf.cast(hypothesis > 0.5 , dtype=tf.float32)
accuracy = tf.reduce_mean(tf.cast(tf.equal(predicted, Y), dtype=tf.float32))

with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    
    for step in range(10001):
        cost_val, _ = sess.run([cost, train], feed_dict={X:x_data, Y:y_data})
        if step % 500 == 0:
            print(step, cost_val)            
    
    h, c, a = sess.run([hypothesis, predicted, accuracy],feed_dict={X:x_data, Y:y_data})    
    print("\nHypothesis: ", h, "\nCorrect (Y): ", c, "\nAccuracy: ",a)
    
```
