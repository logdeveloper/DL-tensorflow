## Neural Network 실습

````python
import tensorflow.compat.v1 as tf 
tf.disable_v2_behavior()
import numpy as np
````

````python
tf.set_random_seed(777)

x_data=np.array([[0,0],[0,1],[1,0],[1,1]], dtype=np.float32)
y_data=np.array([[0],[1],[1],[0]], dtype=np.float32)
````

````python
# 변수 선언
X = tf.placeholder(tf.float32, [None,2]) # x1, x2 
Y = tf.placeholder(tf.float32, [None,1]) # y
W = tf.Variable(tf.random_normal([2,1]), name ="weight") # x변수 2개, y변수 1개
b = tf.Variable(tf.random_normal([1]), name = "bias")
````

````python
# logistic regression
hypothesis = tf.sigmoid(tf.matmul(X,W) + b)

# cost/loss function
cost = -tf.reduce_mean( Y * tf.log(hypothesis) + (1-Y)*tf.log(1-hypothesis))
train = tf.train.GradientDescentOptimizer(learning_rate=0.1).minimize(cost)

# Accuracy computation 
# 만약 hypotesis 결과가 0.5 이상이면 False
predicted = tf.cast(hypothesis > 0.5, dtype=tf.float32)
accuracy = tf.reduce_mean(tf.cast(tf.equal(predicted, Y), dtype=tf.float32))
````

````python
# 그래프 실행
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    
    for step in range(10001):
        _, cost_val, w_val = sess.run(
            [train, cost, W] , feed_dict={X:x_data, Y:y_data}
        )
        if step % 100 == 0:
            print(step, cost_val, w_val)
            
    h, c, a = sess.run(
        [hypothesis, predicted, accuracy], feed_dict={X:x_data, Y: y_data}
    )
    
    print("\nHypotesis: ", h, "\nCorrect: ", c, "\nAccuracy: ", a )
````

위의 regressor으로는 xor 문제가 해결이 안된다. 그래서 나온 개념이 `deep learning`. layer를 늘려서 모델을 구축해보자.

xor nn(nueral network) 적용

````python
tf.set_random_seed(777)

# datset
x_data=np.array([[0,0],[0,1],[1,0],[1,1]], dtype=np.float32)
y_data=np.array([[0],[1],[1],[0]], dtype=np.float32)
````

````python
# 변수 생성
X = tf.placeholder(tf.float32, [None, 2])
Y = tf.placeholder(tf.float32, [None, 1])

##################
#### 달리진곳 ####
##################

# layer의 input 2, ouput 2
W1 = tf.Variable(tf.random_normal([2,2]), name ='weight1')
# output 과 동일하게 2로 
b1 = tf.Variable(tf.random_normal([2]), name = 'bias1') 

layer1 = tf.sigmoid(tf.matmul(X, W1) + b1)

# layer1의 output 2를 그대로 가져와서 W2 input에 넣는다. 
W2 = tf.Variable(tf.random_normal([2,1]), name ='weight1')
# weight의 output 값
b2 = tf.Variable(tf.random_normal([1]), name = 'bias1') 

##################

# 모델 구축
hypothesis = tf.sigmoid(tf.matmul(layer1, W2) + b2)

# cost/loss 함수
cost = -tf.reduce_mean(Y * tf.log(hypothesis)+(1-Y) * tf.log(1-hypothesis))
train = tf.train.GradientDescentOptimizer(learning_rate=0.1).minimize(cost)

# 예측값, 정확도
predicted = tf.cast(hypothesis > 0.5, dtype=tf.float32)
accuracy = tf.reduce_mean(tf.cast(tf.equal(predicted, Y), dtype=tf.float32))
````

````python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    
    for step in range(10001):
        _, cost_val = sess.run([train, cost], feed_dict={X:x_data, Y:y_data})
        if step % 1000 == 0:
            print(step, cost_val)
    h, p, a = sess.run(
        [hypothesis, predicted, accuracy], feed_dict={X:x_data, Y:y_data}
    )
    print(f"\nHypothesis :\n{h} \nPredicted:\n{p} \nAccuracy:\n{a}")
````

- layer를 늘리면 예상하던 대로 10000번의 시도 결과, cost는 점점 작아지고, 정확도는 높아졌다.

- 이젠 layer의 출력 값을 더 많이 늘리면 어떻게 되는지 알아보자.
    - Wide NN for XOR : output 출력을 늘리는 것. ex) W1 = tf.Variable(tf.random_normal([2,10]), name = 'weight1')
    - Deep NN for XOR : layer을 늘려가는 것. ex ) layer1, layer2, ...

````python
# 변수 생성 : W, b, layer
W1 = tf.Variable(tf.random_normal([2,10]), name = 'weight1')
b1 = tf.Variable(tf.random_normal([10]), name = 'bias1')
layer1 = tf.sigmoid(tf.matmul(X,W1) + b1)

W2 = tf.Variable(tf.random_normal([10,10]), name = 'weight2')
b2 = tf.Variable(tf.random_normal([10]), name = 'bias2')
layer2 = tf.sigmoid(tf.matmul(layer1,W2) + b2)

W3 = tf.Variable(tf.random_normal([10,10]), name = 'weight3')
b3 = tf.Variable(tf.random_normal([10]), name = 'bias3')
layer3 = tf.sigmoid(tf.matmul(layer2,W3) + b3)

W4 = tf.Variable(tf.random_normal([10,1]), name = 'weight4')
b4 = tf.Variable(tf.random_normal([1]), name = 'bias4')

# 가설 
hypothesis = tf.sigmoid(tf.matmul(layer3, W4)+b4)
````

````python
# cost/loss 함수
cost = -tf.reduce_mean(Y * tf.log(hypothesis) + (1-Y)*tf.log(1-hypothesis))
train = tf.train.GradientDescentOptimizer(learning_rate=0.1).minimize(cost)

# 예측값,정확도
predicted = tf.cast(hypothesis>0.5, dtype=tf.float32)
accuracy = tf.reduce_mean(tf.cast(tf.equal(predicted, Y), dtype=tf.float32))
````

````python
# 모델 실행
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    
    for step in range(10001):
        _, cost_val = sess.run([train, cost], feed_dict={X:x_data, Y:y_data})
        if step % 1000 == 0:
            print(step, cost_val)
    h, p, a = sess.run(
        [hypothesis, predicted, accuracy], feed_dict={X:x_data, Y:y_data}
    )
    print(f"\nHypothesis :\n{h} \nPredicted:\n{p} \nAccuracy:\n{a}")
````

- 그결과, 작은값은 더 작아지고, 큰 값은 더 커졌다. 
- layer를 더 쌓을 수록 error가 작아진다.

````python
X = tf.placeholder(tf.float32, [None, 2], name='x')
Y = tf.placeholder(tf.float32, [None, 1], name='y')

# 그래프를 그릴때 한번에 보면 복잡하기 때문에, 레이어별로 기능별로 보기 쉽게 해준다.
with tf.name_scope("Layer1"):
    W1 = tf.Variable(tf.random_normal([2,2]), name="weight_1")
    b1 = tf.Variable(tf.random_normal([2]), name="bias_1")
    layer1 = tf.sigmoid(tf.matmul(X, W1) + b1)
    # 여러개의 값을 가질 경우엔 histogram을 사용 / 하나는 scalar 사용
    tf.summary.histogram("W1", W1)
    tf.summary.histogram("b1", b1)
    tf.summary.histogram("Layer1", layer1)
    
with tf.name_scope("Layer2"):
    W2 = tf.Variable(tf.random_normal([2,1]), name="weight_2")
    b2 = tf.Variable(tf.random_normal([1]), name="bias_2")
    hypothesis = tf.sigmoid(tf.matmul(layer1, W2) + b2)
    
    tf.summary.histogram("W2", W2)
    tf.summary.histogram("b2", b2)
    tf.summary.histogram("Hypothesis", hypothesis)    
    
````

````python
with tf.name_scope("Cost"):
    cost = -tf.reduce_mean(Y * tf.log(hypothesis) + (1-Y) * tf.log(1 - hypothesis))
    tf.summary.scalar("Cost", cost)
    
with tf.name_scope("Train"):
    train = tf.train.AdamOptimizer(learning_rate=0.01).minimize(cost)    
````

````python
predicted = tf.cast(hypothesis > 0.5, dtype=tf.float32)
accuracy = tf.reduce_mean(tf.cast(tf.equal(predicted, Y), dtype=tf.float32))
tf.summary.scalar("accuracy", accuracy)
````

forward prop, back prop 를 적용해보자,

````python
x_data = [[1.],
          [2.],
          [3.]]

y_data = [[1.],
          [2.],
          [3.]]

X = tf.placeholder(tf.float32, shape=[None,1])
Y = tf.placeholder(tf.float32, shape=[None,1])
````

````python
# 잘못된 weight 모델을 적용해보자. 
W = tf.Variable(tf.truncated_normal([1,1]))
b = tf.Variable(5.)

# Forward prop 구현
hypothesis = tf.matmul(X, W) + b

# diff 
assert hypothesis.shape.as_list() == Y.shape.as_list() # 모델 적용한 값과 실제 Y값을 비교함.
diff = (hypothesis- Y) # 예측값과 실제값의 차이 
````

````python
# Back prop (chain rule)

d_l1 = diff
d_b = d_l1
d_w = tf.matmul(tf.transpose(X), d_l1)

print(X, W, d_l1, d_w)
````

````python
# Updating network using gradients 
# gradients를 사용하여 network를 update함

learning_rate=0.1
step = [
    tf.assign(W,W-learning_rate * d_w),
    tf.assign(b,b-learning_rate * tf.reduce_mean(d_b)),
]

# cost function
RMSE = tf.reduce_mean(tf.square((Y-hypothesis)))

sess = tf.InteractiveSession()
init = tf.global_variables_initializer()
sess.run(init)

for i in range(1000):
    print(i, sess.run([step, RMSE], feed_dict={X:x_data, Y:y_data}))

print('-------------- y_pred --------------------')
print(sess.run(hypothesis, feed_dict={X:x_data}))
````

## zoo 예제로 NN 적용

````python
xy= np.loadtxt('/Data-Science/alldeaplearning/data/data-04-zoo.csv', 
              delimiter=',', dtype=np.float32)
X_data = xy[:, 0:-1]
N = X_data.shape[0]
y_data = xy[:,[-1]]

print("y has one of the following values")
print(np.unique(y_data)) # 0~6으로 구성
````

````python
# X_data.shape = (101, 16) => 101 samples, 16 features
# y_data.shape = (101, 1)  => 101 samples, 1 label

print("Shape of X data:", X_data.shape)
print("Shape of y data:", y_data.shape)
````

````python
nb_classes = 7 # 0~6

X = tf.placeholder(tf.float32, [None, 16])
y = tf.placeholder(tf.int32, [None, 1]) # 0~6

target = tf.one_hot(y, nb_classes) # float32는 error 발생. int 형만 가능.
target = tf.reshape(target, [-1, nb_classes])
target = tf.cast(target, tf.float32)
````

````python
# W, b 설정
W = tf.Variable(tf.random_normal([16, nb_classes]), name = 'weight')
b = tf.Variable(tf.random_normal([nb_classes]), name = 'bias')
````

````python
#sigmoid function 설정
def sigma(x):
    return 1./ (1. + tf.exp(-x))

def sigma_prime(x):
    return sigma(x) * (1. - sigma(x))

# 앞에서부터 값을 넣어보자! -> Forward propation
layer_1 = tf.matmul(X, W) + b
y_pred = sigma(layer_1)

# Loss Function (end of forward propagtion)
loss_i = - target * tf.log(y_pred) - (1. - target) * tf.log(1. - y_pred)
loss = tf.reduce_sum(loss_i)
````

````python
# Dimension Chekck
assert y_pred.shape.as_list() == target.shape.as_list()
````

````python
# Back prop (chain rule)
d_loss = (y_pred - target) / (y_pred * (1. - y_pred) + 1e-7)
d_sigma = sigma_prime(layer_1)
d_layer = d_loss * d_sigma
d_b = d_layer
d_W = tf.matmul(tf.transpose(X), d_layer)
````

````python
# Updating network using gradient
learning_rater = 0.01
train_step = [
    tf.assign(W, W - learning_rate * d_W),
    tf.assign(b, b - learning_rate * tf.reduce_sum(d_b)),
]
````

````python
# Prediction and Accuracy
prediction = tf.argmax(y_pred, 1)
acct_mat = tf.equal(tf.argmax(y_pred, 1), tf.argmax(target, 1))
acct_res = tf.reduce_mean(tf.cast(acct_mat, tf.float32))
````

````python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    
    for step in range(500):
        sess.run(train_step, feed_dict={X:X_data, y:y_data })
        
        if step % 10 == 0:
            # 아마 300 step 내에서 정확도가 100%로 가는 것을 볼 수 있다. 
            step_loss, acc = sess.run([loss, acct_res], 
                                      feed_dict={X:X_data, y:y_data})
            print("Step : {:5} \t Loss: {:10.5f}\t Acc: {:.2%}".format(
            step, step_loss, acc))
            
    # 자이제 예측 해보자.
    pred  = sess.run(prediction, feed_dict={X: X_data})
    for p, y in zip(pred, y_data):
        msg = "[{}]\t Prediction: {:d}\t True y: {:d}"
        print(msg.format(p == int(y[0]), p, int(y[0])))
````
