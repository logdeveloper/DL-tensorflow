```python
# import tensorflow as tf 
# tensorflow 2.3 version 말고 1.0 version 사용
import tensorflow.compat.v1 as tf 
tf.disable_v2_behavior()
```

#

```python
# x,y train data 셋팅
x_train=[1,2,3]
y_train=[1,2,3]
```

```python
# cost 함수의 변수 설정 
W = tf.Variable(tf.random_normal([1]), name="weight")
b = tf.Variable(tf.random_normal([1]), name="bias")
```

```python
# H = X*W + b
hypothesis =x_train * W + b
```

```python
# cost/lost function
cost = tf.reduce_mean(tf.square(hypothesis-y_train))
cost
```

```python
train = tf.train.GradientDescentOptimizer(learning_rate=0.01).minimize(cost)
```

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    
    for step in range(2001):
        _, cost_val, W_val, b_val = sess.run([train, cost, W, b])
        
        if step % 20 == 0:
            print(step, cost_val, W_val, b_val)
```

```python
# 재생산
tf.set_random_seed(777)
```

```python
W = tf.Variable(tf.random_normal([1]), name="weight")
b = tf.Variable(tf.random_normal([1]), name="bias")
```

```python
# feed_dict를 사용하기 위해 사용되는 placeholder
X = tf.placeholder(tf.float32, shape=[None])
Y = tf.placeholder(tf.float32, shape=[None])
```

```python
# 비용 설정
cost = tf.reduce_mean(tf.square(hypothesis - Y))
```

```python
# 최적화 
train = tf.train.GradientDescentOptimizer(learning_rate=0.01).minimize(cost)
```

```python
# 세션을 통해 그래프 생성
with tf.Session() as sess:
    # Initializes global variables in the graph.
    sess.run(tf.global_variables_initializer())

    # Fit the line
    for step in range(2001):
        _, cost_val, W_val, b_val = sess.run(
            [train, cost, W, b], feed_dict={X: [1, 2, 3], Y: [1, 2, 3]}
        )
        if step % 20 == 0:
            print(step, cost_val, W_val, b_val)

    # Testing our model
    print(sess.run(hypothesis, feed_dict={X: [5]}))
    print(sess.run(hypothesis, feed_dict={X: [2.5]}))
    print(sess.run(hypothesis, feed_dict={X: [1.5, 3.5]}))
```

```python
# 데이터준비
x_train=[1,2,3,4]
y_train=[0,-1,-2,-3]
```

```python
W = tf.Variable([0.3],tf.float32)
b = tf.Variable([-0.3], tf.float32)
```

```python
x=tf.placeholder(tf.float32)
y=tf.placeholder(tf.float32)
```

```python
hypothesis = x*W + b
```

```python
cost = tf.reduce_sum(tf.square(hypothesis - y))
train = tf.train.GradientDescentOptimizer(learning_rate=0.01).minimize(cost)
```

```python
with tf.Session() as sess:
    # Initializes global variables in the graph.
    sess.run(tf.global_variables_initializer())

    # Fit the line
    for step in range(1000):
        sess.run(train, {x:x_train, y : y_train})
        
    W_val, b_val, cost_val = sess.run([W, b, cost], feed_dict={x:x_train, y : y_train})
    print(f"W: {W_val} b:{b_val} cost:{cost_val}")
```
