```python
import tensorflow as tf
import numpy as np
```

```python
# 데이터 셋 
## 방법 1. 직접 입력
x_data = [[73., 80., 75.],
          [93., 88., 93.],
          [89., 91., 90.],
          [96., 98., 100.],
          [73., 66., 70.]]
y_data = [[152.],
          [185.],
          [180.],
          [196.],
          [142.]]

## 방법 2. 파일로 불러와서 데이터 입력
# xy = np.loadtxt('../data-01-test-score.csv', delimiter=',', dtype=np.float32)
# x_data = xy[:, 0:-1]
# y_data = xy[:, [-1]]

tf.model = tf.keras.Sequential()

tf.model.add(tf.keras.layers.Dense(units=1, input_dim=3))
tf.model.add(tf.keras.layers.Activation('linear'))

tf.model.compile(loss='mse', optimizer = tf.keras.optimizers.SGD(lr=1e-5))
tf.model.summary()
history = tf.model.fit(x_data, y_data, epochs=100) # 100번 실행

y_predict = tf.model.predict(np.array([[72., 83., 90.]]))
print( 'y 예측값  : ', y_predict)
```

100번 시도시, loss는 줄어들고 y 예측값 출력도 가능하다.
