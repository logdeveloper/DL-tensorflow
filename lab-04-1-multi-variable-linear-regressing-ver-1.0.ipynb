```python
import tensorflow.compat.v1 as tf 
tf.disable_v2_behavior()
```

```python
tf.set_random_seed(777) # 실습을 위해 임의값 설정, 

# 샘플 데이터 셋 설정
x1_data = [73., 93., 89., 96., 73.]
x2_data = [80., 88., 91., 98., 66.]
x3_data = [75., 93., 90., 100., 70.]

y_data =[152., 185, 180., 196., 142.,] 

# placeholder : 선언과 동시에 초기화 되는 것이 아닌 선언 후 그 다음 값을 전달
x1 = tf.placeholder(tf.float32)
x2 = tf.placeholder(tf.float32)
x3 = tf.placeholder(tf.float32)

Y = tf.placeholder(tf.float32)

w1 = tf.Variable(tf.random_normal([1]), name = 'weight1') # 하나의 값으로 준다.
w2 = tf.Variable(tf.random_normal([1]), name = 'weight2')
w3 = tf.Variable(tf.random_normal([1]), name = 'weight3')
b = tf.Variable(tf.random_normal([1]), name = 'bias')

hypothesis = x1*w1 + x2*w2 + x3*w3 + b

# cost/loss 함수
cost = tf.reduce_mean(tf.square(hypothesis -Y))

optimizer = tf.train.GradientDescentOptimizer(learning_rate=1e-5) # 0.001로 쓰는 것을 1e-5로 표현 
# 가장 작은 learning late를 찾는다.
train = optimizer.minimize(cost)



sess=tf.Session()
sess.run(tf.global_variables_initializer())

for step in range(2001):
    cost_val, hy_val, _ = sess.run([cost, hypothesis, train],
                                  feed_dict={x1:x1_data, x2:x2_data, x3:x3_data, Y:y_data})
    if step % 10 == 0:
        print(step , "Cost: ", cost_val, "\nPrediction:\n", hy_val)
    
```

- 2000번을 시도 했을 때, cost 비용이 줄어 든다. 
- 예측값은  [154.52213 182.23383 181.2558  199.76697 135.95403] 로 실제 값 [152., 185, 180., 196., 142.,] 비슷하다. 
- 하지만 이 방식은 나중에 x가 수백개씩,,.많아 지면 비효율적이기 때문에 주로 metrix를 사용한다.

방법2 : matrix 사용

```python
tf.set_random_seed(777)

# 데이터 셋팅
x_data = [
    [73., 80., 75.],
    [93., 88., 93.],
    [89., 91., 90.],
    [96., 98., 100.],
    [73., 66., 70.]]

y_data = [
    [152.],
    [185.],
    [180.],
    [196.],
    [142.]]

X = tf.placeholder(tf.float32, shape=[None, 3]) # None = N개의 값을 가질 것이다. 값의 갯수를 정해주지 않을때 사용 
Y = tf.placeholder(tf.float32, shape=[None, 1])

# Weight, bias 변수 지정
W = tf.Variable(tf.random_normal([3,1]), name = 'weight')
b = tf.Variable(tf.random_normal([1]), name= 'bias')

# 회귀식 
hypothesis = tf.matmul(X,W) + b

# cost/loss 항수 설정
cost = tf.reduce_mean(tf.square(hypothesis-Y))

# 최소화 
optimizer = tf.train.GradientDescentOptimizer(learning_rate=1e-5)
train = optimizer.minimize(cost)

# session 연결 
sess = tf.Session()
sess.run(tf.global_variables_initializer())

# 실행 
for step in range(2001):
    cost_val, hy_val, _ = sess.run(
        [cost, hypothesis, train], feed_dict={X:x_data, Y:y_data})
    if step %10 == 0 :
        print(step, " Cost: ", cost_val, "\nPrediction\n", hy_val)
```

matrix를 사용할때도 앞의 경우와 마찬가지로 cost가 점점 줄어들고, 예측값은 실제 y 값과 가까워지는 것을 확인할 수 있다.

방법 1. numpy 사용 

```python
import numpy as np

tf.set_random_seed(777)

# numpy 의 단점은 모든 데이터가 같은 타입이어야 함. 
xy = np.loadtxt('../../../DeepLearningZeroToAll/DeepLearningZeroToAll-master/data-01-test-score.csv', delimiter=',', dtype=np.float32)

x_data = xy[:, 0:-1]
y_data = xy[:, [-1]]

print("x_data shape", x_data.shape,"\n",x_data)1
print("y_data shape", y_data.shape,"\n",y_data)
```

```python
X = tf.placeholder(tf.float32, shape=[None, 3])
Y = tf.placeholder(tf.float32, shape=[None, 1])

W = tf.Variable(tf.random_normal([3,1]), name = 'weight')
b = tf.Variable(tf.random_normal([1]), name = 'bias')

hypothesis = tf.matmul(X, W) + b

cost = tf.reduce_mean(tf.square(hypothesis-Y))

optimizer = tf.train.GradientDescentOptimizer(learning_rate=1e-5)
train = optimizer.minimize(cost)
sess = tf.Session()

sess.run(tf.global_variables_initializer())

for step in range(2001):
    cost_val, hy_val, _ = sess.run([cost, hypothesis, train],
                                  feed_dict={X:x_data, Y:y_data})
    
    if step % 100 == 0:
        print(step, "Cost :", cost_val, "\nPrediction\n", hy_val)
```

2000번을 시도했을때 cost는 점점 낮아지고, y 예측값은 실제 y값과 비슷함을 알 수 있다.

방법 2. tensor의 queue runners 이용 

```python
# 1. 파일을 불러옴 (파일 여러개의 리스트도 가능)
filename_queue = tf.train.string_input_producer(
        ['../../../DeepLearningZeroToAll/DeepLearningZeroToAll-master/data-01-test-score.csv'], shuffle=False, name='filename_queue')

# 2. File을 읽어올 Reader을 정의
reader= tf.TextLineReader()
key, value = reader.read(filename_queue)

# 3. value를 읽어와 어떻게 parsing할 것인지 정의
record_defaults = [[0.],[0.],[0.],[0.] ] ## 각각 필드마다 데이터 형식 정의 가능 # 여기선 float으로 정의
xy = tf.decode_csv(value, record_defaults=record_defaults)

# 4. tensorflow batch를 사용하여 데이터를 읽어옴
train_x_batch, train_y_batch = tf.train.batch([xy[0:-1], xy[-1:]], batch_size=10)

# 여기서부턴 동일
X = tf.placeholder(tf.float32, shape=[None, 3])
Y = tf.placeholder(tf.float32, shape=[None, 1])

W = tf.Variable(tf.random_normal([3, 1]), name='weight')
b = tf.Variable(tf.random_normal([1]), name='bias')

hypothesis = tf.matmul(X, W) + b

cost = tf.reduce_mean(tf.square(hypothesis - Y))

optimizer = tf.train.GradientDescentOptimizer(learning_rate=1e-5)

train = optimizer.minimize(cost)

sess = tf.Session()
sess.run(tf.global_variables_initializer())

# session에 연동. 
coord = tf.train.Coordinator()
threads = tf.train.start_queue_runners(sess=sess, coord=coord)

for step in range(2001):
    x_batch, y_batch = sess.run([train_x_batch, train_y_batch]) # 데이터를 펌프해서 가져온다.
    cost_val, hy_val, _ = sess.run(
        [cost, hypothesis, train], feed_dict={X: x_batch, Y: y_batch})
    if step % 500 == 0:
        print(step, "Cost: ", cost_val, "\nPrediction:\n", hy_val)

coord.request_stop()
coord.join(threads)
```
